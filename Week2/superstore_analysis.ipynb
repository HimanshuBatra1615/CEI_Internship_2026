## Step 1: Import Required Libraries

In this step, we import the required Python libraries and establish a SQLite database connection for executing SQL queries.
Load the Superstore dataset into a Pandas DataFrame and inspect its dimensions.

```python
import pandas as pd
import sqlite3
df=pd.read_csv("/content/Sample - Superstore.csv",encoding="latin1")
print("Dataset Shape:",df.shape)
df.head()
```

## Step 2: Create SQL Database

Store the dataset inside a SQLite database to perform SQL operations.

```python
conn = sqlite3.connect("superstore.db")

df.to_sql(
    "superstore",
    conn,
    if_exists="replace",
    index=False
)

print("Table Created Successfully")
```

# Data Exploration
### Explore Table Schema

```python
query = """
PRAGMA table_info(superstore);
"""

pd.read_sql(query, conn)
```

### Display Sample Records

```python
query = """
SELECT *
FROM superstore
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS total_rows
FROM superstore;
"""

pd.read_sql(query, conn)
```

# Data Filtering using WHERE
### Filter Records by Region

```python
query = """
SELECT *
FROM superstore
WHERE Region='West';
"""

pd.read_sql(query, conn)
```

### Filter Records by Category

```python
query = """
SELECT *
FROM superstore
WHERE Category='Technology';
"""

pd.read_sql(query, conn)
```

### Filter Records based on Sales Amount

```python
query = """
SELECT *
FROM superstore
WHERE Sales > 1000;
"""

pd.read_sql(query, conn)
```

# Data Aggregation using GROUP BY
### Total Sales by Region

```python
query = """
SELECT
Region,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Region
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

### Total Sales by Category

```python
query = """
SELECT
Category,
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

### Average Sales and Quantity by Category

```python
query = """
SELECT
Category,
AVG(Sales) AS Average_Sales,
SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY Category;
"""

pd.read_sql(query, conn)
```

# Business Analysis
### Top 10 Selling Products

```python
query = """
SELECT
[Product Name],
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY [Product Name]
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### Top 10 Customers by Sales

```python
query = """
SELECT
[Customer Name],
SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY [Customer Name]
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### Monthly Sales Trend

```python
query = """
SELECT
substr([Order Date],7,4) || '-' ||
substr([Order Date],1,2) AS Month,

SUM(Sales) AS Revenue

FROM superstore

GROUP BY Month

ORDER BY Month;
"""

pd.read_sql(query, conn)
```

# Data Quality Validation
### Duplicate Order Detection

```python
query = """
SELECT
[Order ID],
COUNT(*) AS duplicate_count

FROM superstore

GROUP BY [Order ID]

HAVING COUNT(*) > 1;
"""

pd.read_sql(query, conn)
```

### Missing Value Check

```python
query = """
SELECT COUNT(*) AS Missing_Customers
FROM superstore
WHERE [Customer Name] IS NULL;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Missing_Products
FROM superstore
WHERE [Product Name] IS NULL;
"""

pd.read_sql(query, conn)
```
